```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import warnings
from matplotlib.pyplot import figure as fig
from scipy import stats
sns.set_theme()
np.random.seed(0)  # Set seed

def spearman_plot(rho = 0, samples=10000):
    """
    Determines Spearman correlation plots of a particular
    number of samples given Pearson correlated plots.
    """

    c = np.sqrt(-2*np.log(np.random.rand(samples, 1)))*np.cos(2*np.pi*np.random.rand(samples, 1))
    s = np.sqrt(-2*np.log(np.random.rand(samples, 1)))*np.sin(2*np.pi*np.random.rand(samples, 1))

    # Correlated random samples
    mu_x, mu_y = 0.0, 0.0
    sigma_x, sigma_y = 1, 1

    X = mu_x + sigma_x * c
    Y = mu_y + sigma_y * (rho*c + np.sqrt(1-rho**2)*s)
   
    corr = stats.spearmanr(X, Y).statistic
    corr_po =  '%.2f'%(corr)
    print('Spearman\'s Correlation = ' + str(corr_po))

    fig1, ax1 = plt.subplots()
    ax1.scatter(X, Y, marker=".")
    ax1.set_xlabel('X', fontsize=2)
    ax1.set_ylabel('Y', fontsize=2)
    ax1.set_title('Spearman Correlation = ' + str(corr_po) + ', Pearson Correlation = ' + str(rho))

for rho in [0, 0.2, 0.4, 0.6, 0.8, 1]:
    spearman_plot(rho)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import warnings
from matplotlib.pyplot import figure as fig
from sklearn.feature_selection import mutual_info_regression
sns.set_theme()
np.random.seed(0)  # Set seed

def MI_plot(rho = 0, samples=10000):
    """
    Determines mutual information plots
    given Pearson correlated plots.
    """

    c = np.sqrt(-2*np.log(np.random.rand(samples, 1)))*np.cos(2*np.pi*np.random.rand(samples, 1))
    s = np.sqrt(-2*np.log(np.random.rand(samples, 1)))*np.sin(2*np.pi*np.random.rand(samples, 1))

    # Correlated random samples
    mu_x, mu_y = 0.0, 0.0
    sigma_x, sigma_y = 1, 1

    X = mu_x + sigma_x * c
    Y = mu_y + sigma_y * (rho*c + np.sqrt(1-rho**2)*s)
   
    corr = mutual_info_regression(X,Y)
    corr_po =  '%.2f'%(corr)
    print('Mutual information = ' + str(corr_po))

    fig1, ax1 = plt.subplots()
    ax1.scatter(X, Y, marker=".")
    ax1.set_xlabel('X', fontsize=2)
    ax1.set_ylabel('Y', fontsize=2)
    ax1.set_title('Mutual information = ' + str(corr_po) + ', Pearson Correlation = ' + str(rho))

for rho in [0, 0.2, 0.4, 0.6, 0.8, 1]:
    MI_plot(rho)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import warnings
from matplotlib.pyplot import figure as fig
sns.set_theme()
np.random.seed(0)  # Set seed

def pearson_plot(rho = 0, samples=10000):
    """
    Generates Pearson correlation plots of a particular
    number of samples.
    """

    c = np.sqrt(-2*np.log(np.random.rand(samples, 1)))*np.cos(2*np.pi*np.random.rand(samples, 1))
    s = np.sqrt(-2*np.log(np.random.rand(samples, 1)))*np.sin(2*np.pi*np.random.rand(samples, 1))

    # Correlated random samples
    mu_x, mu_y = 0.0, 0.0
    sigma_x, sigma_y = 1, 1

    X = mu_x + sigma_x * c
    Y = mu_y + sigma_y * (rho*c + np.sqrt(1-rho**2)*s)
   
    corr = (np.mean(X*Y) - np.mean(X)*np.mean(Y))/(np.sqrt(np.mean(X**2) - np.mean(X)**2))/(np.sqrt(np.mean(Y**2) - np.mean(Y)**2))
    corr_po =  '%.2f'%(corr)
    print('Pearson\'s Correlation = ' + str(corr_po))

    fig1, ax1 = plt.subplots()
    ax1.scatter(X, Y, marker=".")
    ax1.set_xlabel('X', fontsize=2)
    ax1.set_ylabel('Y', fontsize=2)
    ax1.set_title('Correlation = ' + str(corr_po))

for rho in [0, 0.2, 0.4, 0.6, 0.8, 1]:
    pearson_plot(rho)
```

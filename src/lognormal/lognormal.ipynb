```python
from scipy.stats import norm
from scipy.stats import lognorm
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure as fig
```

# More on the Gaussian distribution being mistaken as the Lognormal

## Examples of lognormal plots

It is not possible to have a negative heart rate, so we use the lognormal.

```python
def x_norm(mean, sd):
    x = np.linspace(norm.ppf(0.01, loc=mean, scale=sd),
                    norm.ppf(0.99, loc=mean, scale=sd), 200)
    plt.plot(x, norm.pdf(x, loc=mean, scale=sd),
             'r-', lw=2, alpha=0.7, label='norm pdf')
    plt.legend()


def x_lognorm(mean, sd):
    x = np.linspace(lognorm.ppf(0.01, s=sd, scale=np.exp(mean)),
                    lognorm.ppf(0.99, s=sd, scale=np.exp(mean)), 200)
    plt.plot(x, lognorm.pdf(x, s=sd, scale=np.exp(mean)),
             'b-', lw=2, alpha=0.7, label='lognorm pdf')
    plt.legend()


def log_moments(m=1.78, sd=0.0635):
    """
    Draw logmoments.
    """
    lm = np.log((mean**2)/(mean**2 + sd**2)**0.5)
    lsd = np.log(1 + (sd**2)/(mean**2))**0.5
    return lm, lsd


def pri_fig():
    """
    Draw primary figures associated
    with the plots.
    """


def main_pdf(m=1.78, sd=0.0635):
    fig(figsize=(10, 10), dpi=300)
    lm, lsd = log_moments(m, sd)
    x_norm(m, sd)
    x_lognorm(lm, lsd)
```

```python
main_pdf()
```

## Mean = 0.1, variance is high, what happens?

Further, at low variance (and a mean > 6 standard deviations away from 0), the lognormal behaves like a normal.

```python
fig(figsize=(10, 10), dpi=300)


def x_mlognorm(mean, sd):
    x = np.linspace(0, 4, 200)
    lbl = 'lognorm pdf, sd = {s:.2f}'.format(s=float(sd))
    plt.plot(x, lognorm.pdf(x, s=sd, scale=np.exp(mean)), lw=2,
             alpha=0.7, label=lbl, fillstyle='full')
    plt.fill_between(x, lognorm.pdf(x, s=sd, scale=np.exp(mean)),
                     alpha=0.3)
    plt.xlabel("x")
    plt.ylabel("lognorm.pdf(x)")
    plt.legend()


for i in np.linspace(0.1, 2, 5):
    x_mlognorm(mean=0.1, sd=i)
```

## Plots of inverse survival function

Here are some plots of the survival tables.

```python
fig(figsize=(10, 10), dpi=300)


def isf_norm(mean=0, sd=1):
    x = 10**np.linspace(1, 9, 1000)
    lbl = 'Mean = ' + str(mean) + ', SD = ' + str(sd)
    plt.plot(x, norm.isf(1/x, loc=mean, scale=sd), lw=2, alpha=0.7, label=lbl)
    plt.xscale("log")
    plt.legend()
    plt.xlabel("n")
    plt.ylabel("isf(1/n)")
    plt.title("Survival function of normal distribution")


isf_norm()
isf_norm(mean=1.78, sd=0.0635)
```

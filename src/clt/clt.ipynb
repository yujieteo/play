```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure as fig
```

```python
s = np.random.uniform(0,10,10000)
count, bins, ignored = plt.hist(s, 15)
plt.show()
```

```python
a, b = np.random.uniform(0,10,10000), np.random.uniform(0,10,10000)
s = a + b
count, bins, ignored = plt.hist(s, 15)
plt.show()
```

```python
def gaussian_uni(limit=20):
    s = np.random.uniform(0,10,10000)
    for i in range(0, limit):
        a = np.random.uniform(0,10,10000)
        s = s + a
    count, bins, ignored = plt.hist(s, 15)
    plt.show()

gaussian_uni()
```

```python
def bernoulli_uni(limit=20, p=0.5):
    s = np.random.binomial(1, p, 10000)
    for i in range(0, limit):
        a = np.random.binomial(1, p, 10000)
        s = s + a
    count, bins, ignored = plt.hist(s, 15)
    plt.show()

bernoulli_uni(limit=20, p=0.5)
```

```python
def pareto_uni(limit=20):
    s = np.random.pareto(2.5,10000)
    for i in range(0, limit):
        a = np.random.pareto(2.5,10000)
        s = s + a
    count, bins, ignored = plt.hist(s, 15)
    plt.show()

pareto_uni(limit=10000)
```
